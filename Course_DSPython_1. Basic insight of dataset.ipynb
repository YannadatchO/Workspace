## Read data
###### data source = "https://github.com/YannadatchO/Workspace/raw/main/imports-85.data.csv"

```python
import pandas as pd
import numpy as np
```

```python
# Read the online file by the URL provides above, and assign it to variable "df"
url = "https://github.com/YannadatchO/Workspace/raw/main/imports-85.data.csv"
df = pd.read_csv(url, header = None)
df.shape
```

```python
# show the first 5 rows using dataframe.head() method
print(" The first 5 rows of dataframe")
df.head(5)
```

```python
# Write your code below and press Shift+Enter to execute 
print("The last 10 rows of the dataframe\n")
#df.tail(10)
```

## Add Headers

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
```

```python
df.head(5)
```

```python
#we need to replace the "?" symbol with NaN so the dropna() can remove the missing values

df1 = df.replace('?', np.NaN)
```

```python
#we can drop missing values along the column "price" as follows

df =df1.dropna(subset = ["price"], axis = 0)
df.shape
```

```python
df.head(5)
```

## Save data set

```python
#For example, if you would save the dataframe df as automobile.csv to your local machine, you may use the syntax below:

df.to_csv("automobile.csv", index=False)
```
